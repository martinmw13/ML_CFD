# Pipeline exploration template

Build a [`RunConfig`](../src/mlcfd/config/schemas.py) in code, or load YAML. Uncomment execution only when the CSV under `data/` exists.

```python
from pathlib import Path

from mlcfd.config.schemas import (
    DataConfig,
    MeshConfig,
    OutputConfig,
    RunConfig,
    SweepModelConfig,
)
from mlcfd.logging_config import configure_logging
```

```python
configure_logging("INFO")

run = RunConfig(
    mesh=MeshConfig(
        nx=227,
        ny=120,
        lx=17.665369,
        ly=9.0,
        x0=2.66537,
        y0=4.5,
        r=0.5,
    ),
    data=DataConfig(
        input_dir=Path("../data"),
        reynolds_number=50,
    ),
    output=OutputConfig(output_dir=Path("../runs/notebook_pca_smoke")),
    model_name="pca_sklearn",
    model_params=SweepModelConfig(profile="sweep", r_max=8, r_step=2),
)
```
